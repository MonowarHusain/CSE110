```python
#task03
my_list = [10, 1, 20, 3, 6, 2, 5, 11, 15, 2, 12, 14, 17, 18, 29]
for imn in range(0,len(my_list)-1):
    for jkl in range(0,len(my_list)-imn-1):
        if my_list[jkl]<my_list[jkl+1]:
            temp = my_list[jkl]
            my_list[jkl]=my_list[jkl+1]
            my_list[jkl+1]=temp
print(my_list)
```
