```python
#task08
word = input()
result = ''

count = 0
shft = ord('a') - ord('A')
for ch in word:
    if count%2 == 1:
        ch_up = chr(ord(ch) - shft)
        result += ch_up
    count += 1

print(result)
```

```python
word = input()
var_1 = 0
count_1 = 0
while word[var_1] != ',':
    var_1 += 1
count_2 = var_1 + 2
var_2 = len(word)

result = ''
while count_1 < var_1 and count_2 < var_2:
    result += word[count_1] + word[count_2]
    count_1 += 1
    count_2 += 1

if count_1 < var_1:
    result += word[count_1:var_1]
else:
    result += word[count_2:var_2]

print(result)
```
